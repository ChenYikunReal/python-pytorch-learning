```python
import torch

'''
包 torch 包含了多维张量的数据结构以及基于其上的多种数学操作。
另外，它也提供了多种工具，其中一些可以更有效地对张量和任意类型进行序列化。
它有CUDA 的对应实现，可以在NVIDIA GPU上进行张量运算(计算能力>=2.0)。
'''

# 定义一个张量
a = torch.randn(1, 2, 3, 4, 5)
# 测试a对象是否是一个pytorch 张量对象
print(torch.is_tensor(a))
# 测试a对象是否是一个pytorch storage对象
print(torch.is_storage(a))
# 输出a张量中的元素个数
print(torch.numel(a))

# 定义一个张量
a = torch.zeros(4, 4)
# 输出a张量中的元素个数
print(torch.numel(a))

# torch.eye(n, m=None, out=None) n为行 m为列 out为Output tensor 最终返回一个2维张量，对角线位置全1，其它位置全0
print(torch.eye(3))
```
