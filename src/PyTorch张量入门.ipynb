```python
import torch
```

```python
# 检测CUDA
print(torch.cuda.is_available())
```

```python
# 随机生成一个2×2的张量
print(torch.rand(2, 2))
```

```python
# 使用二维列表创建一个2×3的张量
x = torch.tensor([[0, 0, 1], [1, 1, 1], [0, 0, 0]])
print(x)
```

```python
# 使用索引改变张量的一个元素
x[0][1] = 3
print(x)
```

```python
# 生成填充1的张量
one = torch.ones(2, 2)
print(one)
```

```python
# 生成填充0的张量
zero = torch.zeros(2, 2)
print(zero)
```

```python
# 张量进行标准数学运算
print(one + zero)
print(zero - one)
```

```python
# 1×1的张量可以使用item()取出数值
torch.rand(1).item()
```

```python
# 创建CPU中的tensor
cpu_tensor = torch.rand(2)
cpu_tensor.device
```

```python
# 将CPU中的张量复制到GPU中
gpu_tensor = cpu_tensor.to("cuda")
gpu_tensor.device
```

```python
# 找出张量中的最大张量元素
torch.rand(2, 2).max()
```

```python
# 找出张量中的最大元素值
torch.rand(2, 2).max().item()
```

```python
# 生成的是LongTensor
x.type()
```

```python
# 转FloatTensor
float_tensor = x.to(dtype=torch.float32)
float_tensor.type()
```

```python
float_tensor
```

```python
# 求张量以2为底的对数
float_tensor.log2()
```

```python
# 节约内存
random_tensor = torch.rand(2, 2)
random_tensor.log2_()
```

```python
# view()可以完成张量变形
flat_tensor = torch.rand(784)
viewed_tensor = flat_tensor.view(1, 28, 28)
print(viewed_tensor.shape)
```

```python
# reshape()也可以完成张量变形 更为推荐使用
reshaped_tensor = flat_tensor.reshape(1, 28, 28)
print(reshaped_tensor.shape)
```

```python
# 如果元素个数不同会报错
# flat_tensor.view(1, 25, 25)
# flat_tensor.view(3, 28, 28)
```

```python
# 重排张量维度
hwc_tensor = torch.rand(640, 480, 3)
chw_tensor = hwc_tensor.permute(2, 0, 1)
print(chw_tensor.shape)
```
